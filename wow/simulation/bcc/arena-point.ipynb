```python
def calc(rating, coefficient = 1.0):
    # if rating > 1500:
    #     return int(round(1511.26 / (1 + 1639.28 * 2.71828 ** (-0.00412 * rating))))
    # else:
    #     return int(round(0.22 * rating + 14))
    # --
    # foo = 356 + (1157 / (1 + 1675 * 2.71828 ** (-0.00412 * rating)))
    # --
    foo = 580 + (1022 / (1 + 123 * 2.71828 ** (-0.00412 * rating)))
    return int(round(foo * coefficient))
```

```python
from math import inf

def calc2(point, coefficient = 1.0):
    lef, rig = 0, 3000
    if calc(rig) < point:
        return inf
    while lef < rig:
        mid = (lef + rig) >> 1
        if calc(mid, coefficient) >= point:
            rig = mid
        else:
            lef = mid + 1
    return rig
```

```python
import matplotlib.pyplot as plt

def plotRatingToPoint():
    ratings = [i for i in range(100, 3001, 100)]
    points = [calc(i) for i in ratings]

    fig, ax1 = plt.subplots()
    color = 'tab:red'
    ax1.set_xlabel('rating')
    ax1.set_ylabel('point', color = color)
    ax1.plot(ratings, [calc(i, 1.00) for i in ratings], 'ro-')
    ax1.plot(ratings, [calc(i, 0.88) for i in ratings], 'go-')
    ax1.plot(ratings, [calc(i, 0.76) for i in ratings], 'bo-')
    ax1.tick_params(axis = 'y', labelcolor = color)

    fig.suptitle('Rating To Point')
    fig.tight_layout()
    plt.grid()
    plt.show()


plotRatingToPoint()
```

```python
from math import ceil

def plotPointToRating():
    points = [i for i in range(100, 1500, 100)]

    fig, ax1 = plt.subplots()
    color = 'tab:red'
    ax1.set_xlabel('point')
    ax1.set_ylabel('rating', color = color)
    ax1.plot(points, [calc2(i, 1.00) for i in points], 'ro-')
    ax1.plot(points, [calc2(i, 0.88) for i in points], 'go-')
    ax1.plot(points, [calc2(i, 0.76) for i in points], 'bo-')
    ax1.tick_params(axis = 'y', labelcolor = color)

    fig.suptitle('Point To Rating')
    fig.tight_layout()
    plt.grid()
    plt.show()


plotPointToRating()
```
