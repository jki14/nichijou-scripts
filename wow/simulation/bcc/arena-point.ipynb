```python
def calc(rating):
    # if rating > 1500:
    #     return int(round(1511.26 / (1 + 1639.28 * 2.71828 ** (-0.00412 * rating))))
    # else:
    #     return int(round(0.22 * rating + 14))
    return int(round(356 + (1157 / (1 + 1675 * 2.71828 ** (-0.00412 * rating)))))
```

```python
from math import inf

def calc2(point):
    lef, rig = 0, 3000
    if calc(rig) < point:
        return inf
    while lef < rig:
        mid = (lef + rig) >> 1
        if calc(mid) >= point:
            rig = mid
        else:
            lef = mid + 1
    return rig
```

```python
import matplotlib.pyplot as plt

def plotRatingToPoint():
    ratings = [i for i in range(100, 3001, 100)]
    points = [calc(i) for i in ratings]

    fig, ax1 = plt.subplots()
    color = 'tab:red'
    ax1.set_xlabel('rating')
    ax1.set_ylabel('point', color = color)
    ax1.plot(ratings, points, 'ro-')
    ax1.plot(ratings, [round(i * 0.88) for i in points], 'go-')
    ax1.plot(ratings, [round(i * 0.76) for i in points], 'bo-')
    ax1.tick_params(axis = 'y', labelcolor = color)

    fig.suptitle('Rating To Point')
    fig.tight_layout()
    plt.grid()
    plt.show()


plotRatingToPoint()
```

```python
from math import ceil

def plotPointToRating():
    points = [i for i in range(100, 1500, 100)]
    rating5s = [calc2(i) for i in points]
    rating3s = [calc2(ceil(i / 0.88)) for i in points]
    rating2s = [calc2(ceil(i / 0.76)) for i in points]
    
    fig, ax1 = plt.subplots()
    color = 'tab:red'
    ax1.set_xlabel('point')
    ax1.set_ylabel('rating', color = color)
    ax1.plot(points, rating5s, 'ro-')
    ax1.plot(points, rating3s, 'go-')
    ax1.plot(points, rating2s, 'bo-')
    ax1.tick_params(axis = 'y', labelcolor = color)

    fig.suptitle('Point To Rating')
    fig.tight_layout()
    plt.grid()
    plt.show()


plotPointToRating()
```

```python
rating = 2200

print('For %d Rating:' % rating)
print('  5s: %d' % calc(rating))
print('  3s: %d' % round(calc(rating) * 0.88))
print('  2s: %d' % round(calc(rating) * 0.76))
```
